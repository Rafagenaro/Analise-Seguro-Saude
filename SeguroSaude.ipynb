```python
contagem = df['smoker'].value_counts()
sns.barplot(x=contagem.index, y=contagem.values)
plt.title('Contagem de fumantes e não fumantes')
plt.xlabel('Smoker')
plt.ylabel('Quantidade')
plt.show()
```

```python
sns.boxplot(x='smoker', y='charges', data=df)
plt.title('Distribuição de charges por smoker')
plt.xlabel('Smoker')
plt.ylabel('Charges')
plt.show()
```

O boxplot mostra que fumantes têm custos de seguro significativamente mais altos do que não fumantes. A mediana e os quartis dos fumantes estão acima dos não fumantes, e há maior variação, indicando que alguns fumantes possuem custos muito elevados. Portanto, ser fumante aumenta tanto o valor típico quanto a dispersão dos custos do seguro.

```python
corr_matrix = df[['charges', 'age', 'bmi', 'children']].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f')
plt.title('Mapa de Calor da Correlação')
plt.show()
```

```python
sns.scatterplot(x = 'age', y = 'charges', data = df)
plt.title('Relação entre Idade e Custo do Seguro')
plt.xlabel('Idade')
plt.ylabel('Custo do Seguro')
```

O scatterplot mostra que, em geral, quanto mais velha a pessoa, maior o custo do seguro. Mas a relação não é muito forte: dá pra ver que, na mesma idade, algumas pessoas têm custos bem altos e outras bem baixos.
